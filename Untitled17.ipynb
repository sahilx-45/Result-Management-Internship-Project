```python
students={}

def grade(marks):
    if marks>=90:
        return("Grade A")
    elif marks>=80:
        return("Grade B")
    elif marks>=60:
        return("Grade C")
    elif marks>=35:
        return("Grade D")
    else:
        return("Fail")

while True:
    print("1.Add Student :")
    print("2.View Student")
    print("3.Exit")

    choice=input("Enter Your Choice :")
    if choice=="1":
        roll_no=int(input("Enter Roll no :"))
        name=input("Enter Your Name :")

        maths=int(input("Enter marks of Maths:"))
        science=int(input("Enter marks of science:"))
        english=int(input("Enter marks of english:"))
        history=int(input("Enter marks of history:"))

        marks=((maths+science+english+history)/400)*100

        students[roll_no]=[name,marks,grade(marks)]
        print("Student Added......!")

    elif choice=="2":
        roll_no = int(input("Roll No: "))
        if roll_no in students:
            s = students[roll_no]
            print("Name:", s[0])
            print("Marks:", s[1])
            print("Grade:", s[2])
        else:
            print("Student does not exist......")
    elif choice=="3":
        break
    else:
        print("Exit")
    
print(students)
```
